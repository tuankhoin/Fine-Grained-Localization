```python
import pandas as pd
import numpy as np
```

```python
res1 = pd.read_csv('Results/COCK_SIFT_exhaustive_5_3_7_1.csv')
res2 = pd.read_csv('Results/COCK_SIFT_3_5_5_7_1.csv')
```

```python
df = pd.DataFrame({'id':res1['id'], 
                   'x': res1['x']-res2['x'], 
                   'y': res1['y']-res2['y'],
                   'MAE': np.abs(res1['x']-res2['x'])/2 + np.abs(res1['y']-res2['y'])/2,
                   'x1': res1['x'],
                   'y1': res1['y'],
                   'x2': res2['x'],
                   'y2': res2['y']
                   })
df_diff = df.loc[np.abs(df['x'])+np.abs(df['y'])>30]
df_diff.shape
```
