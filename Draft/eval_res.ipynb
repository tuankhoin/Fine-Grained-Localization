```python
df = pd.DataFrame({'id':res1['id'], 
                   'x': res1['x']-res2['x'], 
                   'y': res1['y']-res2['y'],
                   'MAE': np.abs(res1['x']-res2['x'])/2 + np.abs(res1['y']-res2['y'])/2,
                   'x1': res1['x'],
                   'y1': res1['y'],
                   'x2': res2['x'],
                   'y2': res2['y'],
                   #'x3': res3['x'],
                   #'y3': res3['y'],
                   #'x4': res4['x'],
                   #'y4': res4['y']
                   })
df_diff = df.loc[np.abs(df['x'])+np.abs(df['y'])>50]
df_diff.shape
```

```python
dist = lambda p1,p2: np.sum((p1-p2)**2)**0.5

def cluster(r):
    pt1 = r[['x1','y1']].values
    pt2 = r[['x2','y2']].values
    pt3 = r[['x3','y3']].values
    return pt2 if dist(pt1,pt2) < 15 else pt3
```

```python
selected = []
for i in range(len(df)):
    selected.append(cluster(df.iloc[i]))
```

```python
out = pd.DataFrame(selected, columns=['x','y'])
out['id'] = df['id']
out = out[['id','x','y']]
out.to_csv('Results/Hist_Clustered.csv',index=False,header=['id','x','y'])
```
